```
from datetime import datetime
```

```
players["date"] = pd.to_datetime(players["date"])
players = players.sort_values(by="date")
```

```
players = players[players['date']<'2015-01-01']
players.head()
```

```
players.attacking_work_rate.value_counts()
```

```
players['attacking_work_rate'] = players['attacking_work_rate'].replace(['None', 'y', 'stoc', 'norm', 'le'], 'medium')
players.attacking_work_rate.value_counts()
```

```
players.defensive_work_rate.value_counts()
```

```
players['defensive_work_rate'] = players['defensive_work_rate'].replace(['_0', 'o','es', '9','0', '2', 'tocky','1', '7', '5', 'ormal','ean' ], 'medium')
players.defensive_work_rate.value_counts()
```

```
players['attacking_work_rate'].replace({'low':0, 'medium': 1, 'high':2}, inplace = True)
players['defensive_work_rate'].replace({'low':0, 'medium': 1, 'high':2}, inplace = True)
#players.head()
```

```
players.drop(['birthday', 'height', 'weight', 'date', 'preferred_foot'],axis=1, inplace=True)
```

```
players.isnull().sum()
```

```
players.sort_values(by = 'player_id', inplace=True)
players.head()
```

```
goals.sort_values(by = 'player1_id').head()
```

```
player_id = goals.player1_id.unique()
players_goal = pd.DataFrame()
sum_goal = []
for i in player_id:
    sum = 0
    p = goals[goals['player1_id']==i]
    sum = np.sum(p['goal_number'])
    sum_goal.append(sum)
```

```
players_goal = pd.DataFrame({'player_id':player_id, 'goals':sum_goal})
```

```
players_goal.sort_values(by='goals', ascending=False, inplace=True)
players_goal.head()
```

```
columns = []
for i in players.columns:
  columns.append(i)

columns = columns[3:]
```

```
player_id1 = players.player_id.unique()
player_id = []
attacking_work_rate = []
defensive_work_rate = []
players_new = []
for i in player_id1:
    player = players[players['player_id']==i]
    player_id.append(i)
    k = player['attacking_work_rate'].mode()
    attacking_work_rate.append(k)
    defensive_work_rate.append(player['defensive_work_rate'].mode())
    players_new.append(player[columns].mean(skipna = True))
```

```
players_new1 = pd.DataFrame()
players_new1['player_id'] = player_id
players_new1['attacking_work_rate'] = attacking_work_rate
players_new1['defensive_work_rate'] = defensive_work_rate
players_new1[columns] = players_new
players_new1.head()
```

```
players_new1 = players_new1.merge(players_goal, on = 'player_id')
```

```
a = [39,43,48,84,95,119,122,188,222,289,338, 347,439,460,511,542,556,586,648,666,814,838]
for i in range(len(players_new1.attacking_work_rate)):
  if i not in a:
  #if i!=39 and i!=43 and i!=48 and i!=84 and i!=95 and i!=119 and i!=122 and i!=188 and i!=222 and i!=289 and i!=338 and i!=347 and i!=439 and i!=460 and i!=511 and i!=542 and i!=556 and i!=586 and i!=648 and i!=666 and i!=814 and i!=838:
    players_new1.attacking_work_rate[i] = int(players_new1.attacking_work_rate[i])
    #print(i)
  else:
    players_new1.attacking_work_rate[i] = 1
  
```

```
a = [19,43, 86,232, 238, 347, 363, 449, 551, 581, 767]
for i in range(len(players_new1.defensive_work_rate)):
  if i not in a:
    players_new1.defensive_work_rate[i] = int(players_new1.defensive_work_rate[i])
    #print(i)
  else:
    players_new1.defensive_work_rate[i] = 1
```

```
players_new1.head()
```

```
def normalize(data, cols):
  for i in columns:
      data[i] = data[i]/max(data[i])
  return data
```

```
columns = players_new1.columns[1:]
players_new1 = normalize(players_new1, columns)
players_new1.head()
```

```
# CHANGE(UPDATED)

attacking_cols = ['attacking_work_rate', 'crossing', 'finishing', 'heading_accuracy', 'short_passing','volleys',
                  'dribbling', 'curve', 'free_kick_accuracy', 'ball_control', 'acceleration', 'sprint_speed',
                  'agility', 'reactions', 'balance', 'shot_power', 'positioning', 'penalties', 'goals'
                  ]
# setting high weights to those attributes which are essential for attacker and set rest of attributes as zero
weights_attack = [9, 8, 9, 7, 7, 8, 8, 7, 9, 8, 8, 8, 9, 7, 7, 8, 7, 9, 10]
sum_weights_attack = np.sum(weights_attack)

k = 0
k = np.array(players_new1[attacking_cols]*weights_attack)
attacking_score = []
for i in range(len(k)):
  attacking_score.append(np.sum(k[i,:]))

attacking_score = attacking_score/sum_weights_attack
max(attacking_score)
```

```
# set a simple quadratic function as an activation function
def Activation(score):
  score = score*100
  act = score**2/93**2
  return act
```

```
defensive_cols = [ 'heading_accuracy', 'short_passing', 'long_passing', 'ball_control', 'acceleration', 'sprint_speed', 'reactions',
                  'balance', 'shot_power', 'jumping', 'stamina', 'strength', 'long_shots', 'interceptions', 'positioning', 'vision',
                  'marking', 'standing_tackle', 'sliding_tackle'
]
# similarly set high weights to those attributes which makes a player stronger defender and set rest of the attributes as zero
weights_defence = [7, 9, 9, 7, 8, 7, 6, 7, 9, 8, 8, 8, 9, 9, 8, 7, 10, 10, 10]
sum_weights_defence = np.sum(weights_defence)
```

```
k = 0
k = np.array(players_new1[defensive_cols]*weights_defence)

defensive_score = []
for i in range(len(k)):
  defensive_score.append(np.sum(k[i,:]))

defensive_score = defensive_score/sum_weights_defence
max(defensive_score)
```

```
goal_keep_cols = ['reactions', 'balance', 'jumping', 'gk_diving', 'gk_handling', 'gk_kicking', 'gk_positioning', 'gk_reflexes']
# Because of less attributes we need to give very high weights to attributes of goal keeper
weights_gk = [8,8, 9, 10,10,8,9,9]
sum_weights_gk = np.sum(weights_gk)


k = 0
k = np.array(players_new1[goal_keep_cols]*weights_gk)

gk_score = []
for i in range(len(k)):
  gk_score.append(np.sum(k[i,:]))

gk_score = gk_score/sum_weights_gk
max(gk_score)
```

```
attacking_score = Activation(attacking_score)
defensive_score = Activation(defensive_score)
gk_score = Activation(gk_score)
```

```
min(attacking_score)
```

```
players_new1['attacking_score'] = attacking_score
players_new1['defensive_score'] = defensive_score
players_new1['gk_score'] = gk_score

players_new1['score'] = 0
for i in range(len(players_new1)):
  players_new1['score'].iloc[i] = max(players_new1['attacking_score'].iloc[i], players_new1['defensive_score'].iloc[i], players_new1['gk_score'].iloc[i])
```
